```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

df_titan = pd.read_csv(r"C:\Users\Tarun\Desktop\DS\Python\Datasets\Visualization\Titanic_data.csv")
df_titan.head()
```

```python
sn.stripplot(x='Sex',y='Age',data=df_titan)  #similar to scatter and bar
plt.title('Strip plot of sex and age')
plt.show()
```

```python
#hue
sn.stripplot(x='Sex',y='Age',hue='Survived',data=df_titan)
plt.title('Strip plot of sex and age')
plt.show()
```

# Distribution plot

```python
df_ins = pd.read_csv(r"C:\Users\Tarun\Desktop\DS\Python\Datasets\Visualization\insurance_data_with_day.csv")

#similar to histogram
sn.distplot(df_ins['claim'],color='olive')
plt.title('Distribute of insurance claim')
plt.show()
```

```python
sn.distplot(df_ins['claim'],color='olive',kde=False)
plt.title('Distribute of insurance claim')
plt.show()
```

# Joint plot

```python
#similar to scatter and bar

sn.jointplot(x='bmi',y='claim',data=df_ins,color='sandybrown')
plt.title('Joint plot')
plt.show()
```

# Violin plot

```python
#similar to box

sn.violinplot(x='Sex',y='Age',data=df_titan)
plt.title('Violin plot')
plt.show
```

```python
#hue 
sn.violinplot(x='Sex',y='Age',data=df_titan,hue='Survived',split=True)
plt.title('Violin plot')
plt.show
```

# Swarm Plot

```python
#similar to scatter

sn.swarmplot(x='Sex',y='Age',data=df_titan)
plt.title('Swarm plot for sex and age')
plt.show()
```

```python
sn.swarmplot(x='Sex',y='Age',data=df_titan,hue='Survived')
plt.title('Swarm plot for sex and age')
plt.show()
```

# Pair plot

```python
#Combination of histogram and scatterplot
#used to plot univariate and bivariate

df_iris = pd.read_csv(r"C:\Users\Tarun\Desktop\DS\Python\Datasets\Visualization\iris.csv")
df_iris.head()
```

```python
sn.pairplot(df_iris)
plt.show()
```

# Count plot

```python
sn.countplot(x='Sex',data=df_titan)

plt.text(x=-0.1,y=580,s=str(round(df_titan.Sex.value_counts()[0]/len(df_titan)*100,2))+'%')
plt.text(x=0.9,y=320,s=str(round(df_titan.Sex.value_counts()[1]/len(df_titan)*100,2))+'%')

plt.title('Gender wise count of passengers')
plt.show()
```

# Regression plot

```python
sn.regplot(x='sepal_length',y='petal_length',data=df_iris)

plt.title('Regression plot')
plt.show()
```

# Heat map

```python
df_iris.corr()
```

```python
sn.heatmap(data=df_iris.corr(),annot=True,linewidth=0.5,cmap="Blues",cbar=True,
          vmin=-1,vmax=1)
plt.show()
```
