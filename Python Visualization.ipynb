```python
import numpy as np
month = np.arange(1,13)
prices = [101,112,203,304,406,507,608,709,810,795,854,900]

plt.plot(month,prices,color='b',marker="*")
plt.title('Price trend in a year')
plt.xlabel('Month')
plt.ylabel('Prices')
#plt.grid(linestyle='--',linewidth='0.5',color='green')
plt.show()
```

```python
import pandas as pd
df_ins = pd.read_csv(r"C:\Users\Tarun\Desktop\DS\Python\Datasets\Visualization\insurance_data_with_day.csv")
df_ins_groupby_day = df_ins.groupby('dayofmonth')
days = df_ins_groupby_day['dayofmonth'].unique()

plt.plot(days, df_ins_groupby_day["age"].mean(),color="limegreen",marker="*",label="Age")
plt.plot(days, df_ins_groupby_day["bmi"].mean(),color="crimson",marker="*",label="BMI")
plt.plot(days, df_ins_groupby_day["bloodpressure"].mean(),color="goldenrod",marker="*",label="Blood Pressure")

plt.title("Multiline plots")
plt.xlabel("Days")
plt.ylabel("Measure")
plt.legend(loc="center right")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(7,4))
plt.style.use('ggplot')

df_ins_groupby_region_gender = df_ins.groupby(['dayofmonth','gender'])
df_ins_groupby_region_gender.mean()["claim"].plot(ax=ax)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(7,4))
plt.style.use('ggplot')

df_ins_groupby_region_gender = df_ins.groupby(['dayofmonth','gender'])
df_ins_groupby_region_gender.mean()["claim"].unstack().plot(ax=ax)
plt.show()
```

```python
Python_marks=(50,65,40,35,77)
R_marks=(55,72,94,70,85)

index=np.arange(5)

plt.bar(x=index,height=Python_marks,width=0.35,label='Python')
plt.bar(x=index + 0.35,height=R_marks,width=0.35,label="R")

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Scores of Students")
plt.xticks(ticks=index+0.35/2,labels=('A','B','C','D','E'))

plt.legend()
plt.show()
```

```python
Python_marks=(50,65,40,35,77)
R_marks=(55,72,94,70,85)

index=np.arange(5)

plt.bar(x=index,height=Python_marks,width=0.35,label='Python')
plt.bar(x=index,height=R_marks,bottom=Python_marks,width=0.35,label="R")

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Scores of Students")
plt.xticks(ticks=index,labels=('A','B','C','D','E'))

plt.legend(loc="upper left")

plt.show()
```

# Pie plot

```python
from matplotlib import cm

region=df_ins['region'].value_counts()

keys = region.keys().to_list()
print(keys)

counts=region.to_list()
print(counts)

cs=cm.Set1([2,4,6,8])

plt.pie(x=counts,labels=keys,autopct='%1.1f%%',colors=cs)
plt.title('Region wise distribution')
plt.show()
```

```python
#Exploded pie plot

from matplotlib import cm

region=df_ins['region'].value_counts()

keys = region.keys().to_list()
counts=region.to_list()


explode=(0.1,0,0,0)

plt.pie(x=counts,labels=keys,autopct='%1.1f%%',explode=explode)
plt.title('Region wise distribution')
plt.show()
```

```python
#Doughnut pieplot

from matplotlib import cm

region=df_ins['region'].value_counts()

keys = region.keys().to_list()
counts=region.to_list()

plt.pie(x=counts,labels=keys,autopct='%1.1f%%')

circle=plt.Circle(xy=(0,0),radius=0.4,color="white")
plt.gcf()
plt.gca().add_artist(circle)

plt.title('Region wise distribution')
plt.show()
```
