## Ranking in DF

```python
data = { 'Name' : ['Tarun','Tejas','Vandana','Yamini','Vathsa','Bunty','Sanjay','Kundal'],
       'Verbal' : [151,152,151,156,100,101,155,104],
       'Quants' : [158,138,140,147,110,103,129,102],
       'Qualify' : ['Yes','Yes','Yes','Yes','No','No','Yes','No']
       }
df_score = pd.DataFrame(data)
df_score
```

```python
df_score['Verbal_Rank']=df_score.Verbal.rank()
df_score
```

```python
df_score['Verbal_Rank']=df_score.Verbal.rank(method='min')
df_score
```

```python
df_score['Verbal_Rank']=df_score.Verbal.rank(method='max')
df_score
```

```python
# doesnot skip a rank like min or max

df_score['Verbal_Rank']=df_score.Verbal.rank(method='dense')
df_score.sort_values('Verbal_Rank)
```
